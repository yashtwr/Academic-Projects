```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random as rand

rand.seed(80)
```

```python
# Initialize input cosisiting of 24 colors
# Shade of Red:dark red, red, salmon, orange red
# Shade of Green: green yellow, lime, spring green, light sea green
# Shade of Blue: corn flower blue,light sky blue, navy, blue
# Shade of Yellow:vgold, golden red, khaki, yellow
# Shade of Teal: teal, cyan,turquoise, aqua marine
# Shade of Pink:dark magenta, magenta, deep pink, pink

def input_generator():
    input_sample=[(139,0,0), (255,0,0), (250,128,114), (255,69,0), (173,255,47),(0,255,0), (0,255,127), (32,178,170), (100,149,237), 
                  (135,206,250), (0,0,128), (0,0,255), (255,215,0), (218,165,32), (240,230,140), (255,255,0), (0,128,128), (0,255,255),
                  (64,224,208), (127,255,212), (139,0,139), (255,0,255), (255,20,147), (255,192,203)]
    output=[]
    
    # Normalize the inputs to 0,1 range using min-max Normalization
    for elements in input_sample:
        temp=[]
        for i in range(0,3):
            norm = elements[i]/255
            temp.append(norm)         
        output.append(tuple(temp))
    return output
```

```python
# Generating normalised input dataset
input_sample = input_generator()
```

```python
# Initialize array for weights
weights = np.empty([100,100,3])

for i in range(0,100):
    for j in range(0,100):
        for m in range(0,3):
            weights[i][j][m] = rand.uniform(0,1)

plt.imshow(weights)
plt.show
```

```python
#### get_winning_node ####
# Get winning nodes for a given input sample by computing the distance of each weights from given input'=
# Returns the location of winner node 

def get_winning_node(input_data,weights):
    min_distance = -1
    location = [-1,-1]
    for i in range(100):
        for j in range(100):
            distance = np.linalg.norm(input_data - weights[i][j])
            if(i == 0 and j == 0):
                min_distance = distance
                location = [i,j]
            elif(distance < min_distance):
                min_distance = distance
                location = [i,j]
    return location


#### update_weights ####
# Update weight for the winning nodes neighourhood 
# Identify the winner nodes neighbourhood Nc using concentric squares 
# Update weights for the Nc only 
# Returns updated weights 

def update_weights(input_color,winning_node,weights,sigma_k,learning_rate):
    x = winning_node[0]
    y = winning_node[1]
    
    i_start = x - int(sigma_k) if ((x - sigma_k) >= 0) else 0
    i_stop = x + int(sigma_k)  if ((x + sigma_k) <= 99) else 99
    j_start = y - int(sigma_k) if ((y - sigma_k) >= 0) else 0
    j_stop = y + int(sigma_k)  if ((y + sigma_k) <= 99) else 99
    #print(winning_node[0],winning_node[1],i_start,i_stop,j_start,j_stop)
    for i in range(j_start,j_stop+1):
        for j in range(i_start,i_stop+1):
            n_dist = np.linalg.norm(weights[x][y] - weights[i][j])
            Nc = np.exp((-n_dist**2)/(2*sigma_k**2))
            
            for k in range(0,3):
                weights[i][j][k] += learning_rate * Nc * (input_color[k] - weights[i][j][k])
            #print("weights are: ", weights)
    return weights
  

#### train ####
# Run the training unitl epoch = either 20, 40, 100 or 1000
# 
# Returns updated weights 
def train(input_data,weights,alpha,sigma,epoch):
    for k in range(0,epoch):
        #print("The current epoch is:",k)
        # Learning rate decay function: To reduce the learning rate
        learning_rate = alpha * np.exp(-k/epoch)
        
        # Sigma decay function: To reduce the sigma neighbourhood radius parameter                                                                                                                         
        sigma_k = sigma * np.exp(-k/epoch)
        #sigma_k = np.round(sigma_k)
        #print("new value for sigma is: ", sigma_k)
        
        # Iterate for each input color out of 24 colors
        for colour in input_data: 
            winning_node_location = get_winning_node(colour, weights)
            updated_weights = update_weights(colour, winning_node_location, weights, sigma_k, learning_rate)
         
    return updated_weights   


def plot(color_mapping ,sigma):
    fig = plt.figure(figsize=(15, 15))
    rows = 1
    columns = 4   
    print("The plots for sigma = {} is". format(sigma))

    # Adds a subplot at the 1st position
    fig.add_subplot(rows, columns, 1)
  
    # show color map for 20 epoch
    plt.imshow(color_mapping[0])
    plt.axis('off')
    plt.title("After 20 epoch")
  
    # Adds a subplot at the 2nd position
    fig.add_subplot(rows, columns, 2)
  
    # show color map for 40 epoch
    plt.imshow(color_mapping[1])
    plt.axis('off')
    plt.title("After 40 epoch")

    # Adds a subplot at the 3rd position
    fig.add_subplot(rows, columns, 3)

    # show color map for 100 epoch
    plt.imshow(color_mapping[2])
    plt.axis('off')
    plt.title("After 100 epoch")

    # Adds a subplot at the 4th position
    fig.add_subplot(rows, columns, 4)

    # show color map for 1000 epoch
    plt.imshow(color_mapping[3])
    plt.axis('off')
    plt.title("After 1000 epoch")
```

```python
alpha = 0.8
# Change the sigma values here
sigma_list = [1, 10, 30, 50, 70]
epoch_list = [20, 40, 100, 1000]
#epoch_list = [20, 40] #, 100, 1000]
#sigma_list = [50]


for sigma in sigma_list:
    color_map = np.empty([4,100,100,3])
    map_indx = 0
    for epoch in epoch_list:
        color_map[map_indx] = train(input_sample,weights,alpha,sigma,epoch)
        #print("index", map_indx)
        map_indx += 1 
    plot(color_map, sigma)
```
