```python
import tensorflow as tf
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split 
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Loading cleaned and normalized dataset
load_dataset = pd.read_csv("cleaned_normalized_coviddata.csv")
df = load_dataset.iloc[:,3:]
# Splitting the data into input and output values
x=df.iloc[:,:-3]
y=df.iloc[:,-3:].astype('int')
```

```python
# Splitting the data in training and testing dataset
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.8,random_state = 100)
```

```python
x_train = np.array(x_train)
x_train1 = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
```

```python
x_test = np.array(x_test)
x_test1 = np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
```

```python
%store x_train x_train1 y_train
```

### Modifying DNN Architecture

### LSTM Model

### Label: Recovered

```python
tf.keras.backend.clear_session()
lstm_recovered = tf.keras.Sequential([
                                tf.keras.layers.LSTM(50,dropout=0.2,recurrent_dropout=0.2,input_shape=(11,1)),
                                tf.keras.layers.Dense(64,activation = 'relu'),
                                tf.keras.layers.Dense(128,activation = 'relu'),
                                tf.keras.layers.Dense(1,activation = 'sigmoid')])
lstm_recovered.summary()
```

```python
lstm_recovered.compile(optimizer='adam',
              loss='BinaryCrossentropy',
              metrics=['accuracy'])
```

```python
lstm_recovered.fit(x_train1,y_train.iloc[:,-1],epochs=200,validation_split=0.2)
```

```python
lstm_recovered.evaluate(x_test1,y_test.iloc[:,-1])
```

```python
predicted_recovered = lstm_recovered.predict(x_test1) > 0.5
```

```python
predicted_recovered.astype('int')
```

### Label: Deaths

```python
tf.keras.backend.clear_session()
lstm_deaths = tf.keras.Sequential([
                                tf.keras.layers.LSTM(50,dropout=0.2,recurrent_dropout=0.2,input_shape=(11,1)),
                                tf.keras.layers.Dense(128,activation = 'relu'),
                                tf.keras.layers.Dense(128,activation = 'relu'),
                                tf.keras.layers.Dense(1,activation = 'sigmoid')])
lstm_deaths.summary()
```

```python
lstm_deaths.compile(optimizer='adam',
              loss='BinaryCrossentropy',
              metrics=['accuracy'])
```

```python
lstm_deaths.fit(x_train1,y_train.iloc[:,-2],epochs=200,validation_split=0.2)
```

```python
lstm_deaths.evaluate(x_test1,y_test.iloc[:,-2])
```

```python
predicted_deaths = (lstm_deaths.predict(x_test1) > 0.50).astype('int')
```

```python
for i in predicted_deaths:
    print(i)
```

```python
accuracy_score(y_test.iloc[:,-2],predicted_deaths)
```
